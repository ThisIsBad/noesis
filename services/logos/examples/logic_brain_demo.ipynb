This notebook shows basic parser, verifier, and interactive session usage.

```python
from logic_brain import verify, is_tautology, are_equivalent
```

```python
result = verify("P -> Q, P |- Q")
result.valid, result.rule
```

```python
is_tautology("P | ~P").valid, are_equivalent("P -> Q", "~Q -> ~P").valid
```

```python
from logic_brain import Z3Session
s = Z3Session(track_unsat_core=True)
s.declare("x", "Int")
s.assert_constraint("x > 0", name="positive")
s.assert_constraint("x < 0", name="negative")
check = s.check()
check.status, check.error_type, check.suggestions
```
